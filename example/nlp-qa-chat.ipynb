````python
import json

with open('../2022/onet-qa-sci-2022p6-20.json', mode='r', encoding='utf-8') as fo:
    corpus = json.loads(fo.read())

corpus
````

````python
print(len(corpus))
````

````python
# Create Question for chat
cat_word = {
    'table': 'กำหนดตาราง',
    'list': '',
    'image-desc': 'ใช้คำบรรยายรูปภาพต่อไปนี้เพื่อตอบคำถาม',
    'text-info': '',
    '':''
}

for qidx, qi in enumerate(corpus[:]):
    header = ""
    materials = qi.get('material', [])
    question_text = qi.get('question', '')
    question_id = qi.get('id', 0)
    choices = qi.get('choice', [])
    for mi in materials:
        mat_type = mi.get('type', '')
        mat_data = mi.get('data', [])

        if mat_type == 'table':
            header += f'{cat_word["table"]}\n'
            header += f'{mat_data}\n'
        else:
            for md in mat_data:
                header += md + '\n'
    header += f'{question_text}\n'

    print(f'คำถามข้อที่ {question_id} {header}')
    for ii, ci in enumerate(choices):
        print(f'{ii+1}. {ci}')

    print('-'*50)
````
